```python
df=pd.read_csv('https://raw.githubusercontent.com/nagarjuna843/test/main/Player_Details.xlsx%20-%20Backend%20Data.csv',header=[1]) # Reading the first row as header
df=df.drop(['Unnamed: 0'],axis='columns') # Dropping the forst row
df.head(5) # Extrcating the sample 5 rows of data
```

```python
from pandasql import sqldf
output = sqldf("select strftime('%m',Signup_Date) as month,strftime('%Y',Signup_Date) as year,COUNT(DISTINCT(Src_Player_Id)) from df GROUP BY 1,2 ORDER BY 1,2 DESC")
output # Extracting the unique customer count data
```

```python
from pandasql import sqldf
output = sqldf("select strftime('%Y',Signup_Date) as month,COUNT(DISTINCT(Src_Player_Id)) AS customers from df GROUP BY 1")
output #Extracting the unique customer data over the years
```
